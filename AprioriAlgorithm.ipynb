```python
#Her satırda tek bir gözlem birimi varmış gibiydi bunu virgül ile tek tek ayırma işlemi yaptık.
#Ürünleri eşsiz olacak şekilde değerlendirmemiz gerekiyor.

data = list(df["1,3,4"].apply(lambda x:x.split(',')))
data
```

```python
# TransactionEncoded bize istediğimiz True-False arrayını dönüştürmede yardımcı oluyor.

from mlxtend.preprocessing import TransactionEncoder
```

```python
#Veri setini istenilen True-False array’ine çevirdim.

te = TransactionEncoder()
te_data = te.fit(data).transform(data)
df = pd.DataFrame(te_data, columns = te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
# df = True-False array ile dönüştürülmüş dataframe
# min_support = Tüm kombinasyonların support değerini istemiyorum bu yüzden belirli bir eşik değerini supporta göre getir.
# use_colnames = True, sutün isimlerini göster.
# verbose = 1, Toplam kombinasyon sayısını verir. 
```

```python
deger= input(' Min. Support değerini girin: ')
min_support = float(deger)
```

```python
df1 = apriori(df, min_support, use_colnames = True, verbose=1)
df1
```

```python
dosya = "C:\\Users\\Administrator\\Desktop\\sonuc_datasetName_minSup.csv"

df1.to_csv(dosya)
```

```python
from mlxtend.frequent_patterns import association_rules

association_rules(df1, metric = "confidence", min_threshold = 0.6)
```

```python
#Örneğin; support değeri %50 ve confidence değeri %75 diyerek dataframe üzerinden indirgemeler yapabiliriz.
```

```python
rules = association_rules(df1, metric = "confidence", min_threshold = 0.6)
rules[ (rules['confidence'] >= 0.6) & (rules['support'] >= 0.75) ]
```
